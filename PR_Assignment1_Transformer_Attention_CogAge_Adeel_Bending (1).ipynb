```python
import pandas as pd
import numpy as np
from pathlib import Path

sample_file = Path("/content/drive/MyDrive/CogAge/sensory_data_Atomic_Activities/HAR_PR_dataset_ver2/examples/Adeel_Bending_1_extracted/PhoneAccelerometer_Event_0.data")

df = pd.read_csv(sample_file, sep=";", header=None)

print("File loaded:", sample_file.name)
print("Raw dataframe shape:", df.shape)

X = df.iloc[:, 1:].values.astype(np.float32)

print("X shape:", X.shape)
print("\nFirst 5 rows of X:")
print(X[:5])
```

Normalize X

```python
X_raw = X.copy()

mean = X_raw.mean(axis=0)
std = X_raw.std(axis=0)

X_attn = (X_raw - mean) / (std + 1e-8)

print("Original X shape:", X_raw.shape)
print("Normalized X_attn shape:", X_attn.shape)

print("\nOriginal feature means:")
print(mean)

print("\nOriginal feature std:")
print(std)

print("\nAfter normalization mean approx:")
print(X_attn.mean(axis=0))

print("\nAfter normalization std approx:")
print(X_attn.std(axis=0))

print("\nFirst 5 rows of normalized X_attn:")
print(X_attn[:5])
```

```python
import numpy as np

def stable_softmax(matrix, axis=1):
    matrix_shifted = matrix - np.max(matrix, axis=axis, keepdims=True)
    exp_matrix = np.exp(matrix_shifted)
    return exp_matrix / np.sum(exp_matrix, axis=axis, keepdims=True)

# Level 1: Direct self-attention
level1_scores = np.dot(X_attn, X_attn.T)

level1_weights = stable_softmax(level1_scores, axis=1)

level1_output = np.dot(level1_weights, X_attn)

print("Input X_attn shape:", X_attn.shape)
print("Level 1 scores shape:", level1_scores.shape)
print("Level 1 attention weights shape:", level1_weights.shape)
print("Level 1 output shape:", level1_output.shape)

print("\nCheck first 5 row sums of attention weights:")
print(level1_weights[:5].sum(axis=1))

print("\nFirst 5 rows of Level 1 output:")
print(level1_output[:5])
```

```python
import numpy as np

# Level 2: Linear projection
np.random.seed(42)

INPUT_DIM = X_attn.shape[1]   # 3 sensor features: x, y, z
HIDDEN_DIM = 8                # projected feature dimension

U = np.random.randn(INPUT_DIM, HIDDEN_DIM).astype(np.float32) * 0.01

X_prime = np.dot(X_attn, U)

print("Input X_attn shape:", X_attn.shape)
print("Weight matrix U shape:", U.shape)
print("Projected X_prime shape:", X_prime.shape)

print("\nFirst 5 rows of U:")
print(U[:5])

print("\nFirst 5 rows of X_prime:")
print(X_prime[:5])
```

```python
# Level 2: Attention on projected X_prime

level2_scores = np.dot(X_prime, X_prime.T) / np.sqrt(HIDDEN_DIM)

level2_weights = stable_softmax(level2_scores, axis=1)

level2_output = np.dot(level2_weights, X_attn)

print("X_prime shape:", X_prime.shape)
print("Level 2 scores shape:", level2_scores.shape)
print("Level 2 attention weights shape:", level2_weights.shape)
print("Level 2 output shape:", level2_output.shape)

print("\nCheck first 5 row sums of Level 2 attention weights:")
print(level2_weights[:5].sum(axis=1))

print("\nFirst 5 rows of Level 2 output:")
print(level2_output[:5])
```

```python
import os
import json
import numpy as np

np.save(os.path.join(SAVE_ROOT, "level2_U_weight_matrix.npy"), U)
np.save(os.path.join(SAVE_ROOT, "level2_X_prime.npy"), X_prime)
np.save(os.path.join(SAVE_ROOT, "level2_scores.npy"), level2_scores)
np.save(os.path.join(SAVE_ROOT, "level2_attention_weights.npy"), level2_weights)
np.save(os.path.join(SAVE_ROOT, "level2_output.npy"), level2_output)

metadata_level2 = {
    "level2_formula": "Y = softmax((X_prime X_prime^T) / sqrt(d)) X",
    "projection_formula": "X_prime = X U",
    "input_X_shape": list(X_attn.shape),
    "U_shape": list(U.shape),
    "X_prime_shape": list(X_prime.shape),
    "hidden_dim": HIDDEN_DIM,
    "level2_scores_shape": list(level2_scores.shape),
    "level2_attention_weights_shape": list(level2_weights.shape),
    "level2_output_shape": list(level2_output.shape)
}

with open(os.path.join(SAVE_ROOT, "metadata_level2.json"), "w") as f:
    json.dump(metadata_level2, f, indent=4)

print("Level 2 files saved successfully.")
print("Save folder:")
print(SAVE_ROOT)

print("\nSaved files:")
for item in os.listdir(SAVE_ROOT):
    print(item)
```

```python
# Level 3: Create Query, Key, Value matrices

np.random.seed(42)

D_MODEL = X_attn.shape[1]   # 3 input features
D_K = 8                     # query/key/value hidden dimension

W_Q = np.random.randn(D_MODEL, D_K).astype(np.float32) * 0.01
W_K = np.random.randn(D_MODEL, D_K).astype(np.float32) * 0.01
W_V = np.random.randn(D_MODEL, D_K).astype(np.float32) * 0.01

Q = np.dot(X_attn, W_Q)
K = np.dot(X_attn, W_K)
V = np.dot(X_attn, W_V)

print("Input X_attn shape:", X_attn.shape)

print("\nW_Q shape:", W_Q.shape)
print("W_K shape:", W_K.shape)
print("W_V shape:", W_V.shape)

print("\nQ shape:", Q.shape)
print("K shape:", K.shape)
print("V shape:", V.shape)

print("\nFirst 5 rows of Q:")
print(Q[:5])

print("\nFirst 5 rows of K:")
print(K[:5])

print("\nFirst 5 rows of V:")
print(V[:5])
```

```python
# Level 3: Scaled Dot-Product Attention

level3_scores = np.dot(Q, K.T) / np.sqrt(D_K)

level3_weights = stable_softmax(level3_scores, axis=1)

level3_output = np.dot(level3_weights, V)

print("Q shape:", Q.shape)
print("K shape:", K.shape)
print("V shape:", V.shape)

print("\nLevel 3 scores shape:", level3_scores.shape)
print("Level 3 attention weights shape:", level3_weights.shape)
print("Level 3 output shape:", level3_output.shape)

print("\nCheck first 5 row sums of Level 3 attention weights:")
print(level3_weights[:5].sum(axis=1))

print("\nFirst 5 rows of Level 3 output:")
print(level3_output[:5])
```
